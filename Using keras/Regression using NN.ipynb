It contains model-specific fuel consumption ratings and estimated carbon dioxide emissions for new light-duty vehicles.
Here we will try to estimate CO2 emission for new vehicle.

```python
import tensorflow as tf
print("TensorFlow version: {}".format(tf.__version__))
print("Keras version: {}".format(tf.keras.__version__))
```

```python
import pandas as pd 
```

```python
df = pd.read_csv('FuelConsumption.csv',engine = 'python')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
final_df = df[['ENGINESIZE', 'CYLINDERS', 'FUELTYPE', 'FUELCONSUMPTION_CITY',
                   'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB',
                   'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
```

```python
print(final_df.shape)
final_df.head()
```

```python
final_data = pd.get_dummies(final_df)
```

```python
final_data.head()
```

```python
final_data.shape
```

```python
X = final_data.drop(['CO2EMISSIONS'], axis = 1)
y = final_data['CO2EMISSIONS']
```

```python
print(X.shape)
X.head()
```

```python
X.iloc[3,:]
```

```python
print(y.shape)
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
print(X_train.shape, y_train.shape,'\n', X_test.shape, y_test.shape)
```

Start prediction Using NN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
```

```python
# create model (Without Dropout it gives better result)

model = Sequential()
model.add(Dense(64, input_dim=10, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='relu'))

optimizer = tf.keras.optimizers.RMSprop(0.001)

# Compile model
model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
```

```python
model.summary()
```

```python
model.fit( X_train, y_train, epochs=500, validation_split = 0.2, batch_size=64)
```

```python
model.evaluate(X_test, y_test)
```
